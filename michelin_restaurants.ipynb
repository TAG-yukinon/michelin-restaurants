```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

df = pd.read_csv('drive/MyDrive/michelin_my_maps.csv')
```

```python
df.drop_duplicates().reset_index(drop=True)
```

```python
df.rename(columns={"WebsiteUrl": "WebUrl"}, inplace=True)
df.head(5)
```

```python
df['MaxPrice'] = df['MaxPrice'].str.replace(',', '')
df['MinPrice'] = df['MinPrice'].str.replace(',', '')

df['MaxPrice'] = df['MaxPrice'].fillna(df['MaxPrice'].median())
df['MinPrice'] = df['MinPrice'].fillna(df['MinPrice'].median())

df['MaxPrice'] = df['MaxPrice'].astype(int)
df['MinPrice'] = df['MinPrice'].astype(int)

df
```

```python
boxplot = df.boxplot('Longitude')
```

```python
pie = df.Award.value_counts().plot(kind='pie')
```

```python
heatmap = sns.heatmap(df.corr(), annot=True)
```

```python
fig, ax = plt.subplots(figsize=(12, 9))

countplot = sns.countplot(ax=ax, data=df, x="Currency", hue="Award")
```

```python
count_nan_in_df = df.isnull().sum()
print (count_nan_in_df)
```

```python
df['Currency'] = df['Currency'].fillna(df['Currency'].mode()[0])

df['PhoneNumber'] = df['PhoneNumber'].fillna(df['PhoneNumber'].mean())

df['WebUrl'] = df['WebUrl'].fillna(df['WebUrl'].mode()[0])

df['FacilitiesAndServices'] = df['FacilitiesAndServices'].fillna(df['FacilitiesAndServices'].mode()[0])
```

```python
count_nan_in_df = df.isnull().sum()
print (count_nan_in_df)
```

```python
k2, p = stats.normaltest(df['MaxPrice'].sample(n=200))

alpha = 1e-3

print("p = {:g}".format(p))

if p < alpha: 
    print("The null hypothesis can be rejected")

else:
    print("The null hypothesis cannot be rejected")
```
